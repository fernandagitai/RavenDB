```python
import pandas as pd
```

```python
df = pd.read_csv('../dados/Fifa_world_cup_matches.csv')
df.head(10)
```

# Tabela SeleÃ§Ãµes

```python
lista_selecoes = list(df['team1'].unique())
lista_selecoes.extend(list(df['team2'].unique()))

df_selecao = {"nome": [], "pontos": [], "vitorias": [], "empates": [], "derrotas": [], "gols_pro": [], "gols_contra": [], "grupo": [], "saldo_gols": [], "aproveitamento": []}	

for selecao in set(lista_selecoes):

    grupos = df[(df['team1'] == selecao) | (df['team2'] == selecao)]['category'].unique()

    for grupo in grupos:
        df_grupo = df[df['category'] == grupo]
        pontos = 0
        vitorias = 0
        empates = 0
        derrotas = 0
        gols_pro = 0
        gols_contra = 0

        df_aux = df_grupo.loc[(df['team1'] == selecao)] # time da casa
        if len(df_aux) > 0:
            for index, row in df_aux.iterrows():
                if row['number of goals team1'] > row['number of goals team2']:
                    pontos += 3
                    vitorias += 1
                elif row['number of goals team1'] == row['number of goals team2']:
                    pontos += 1
                    empates += 1
                else:
                    derrotas += 1

                gols_pro += row['number of goals team1']
                gols_contra += row['number of goals team2']

        df_aux = df_grupo.loc[(df['team2'] == selecao)] # time visitante
        if len(df_aux) > 0:
            for index, row in df_aux.iterrows():
                if row['number of goals team2'] > row['number of goals team1']:
                    pontos += 3
                    vitorias += 1
                elif row['number of goals team2'] == row['number of goals team1']:
                    pontos += 1
                    empates += 1
                else:
                    derrotas += 1

                gols_pro += row['number of goals team2']
                gols_contra += row['number of goals team1']


        df_selecao["nome"].append(selecao)
        df_selecao["pontos"].append(pontos)
        df_selecao["vitorias"].append(vitorias)
        df_selecao["empates"].append(empates)
        df_selecao["derrotas"].append(derrotas)
        df_selecao["gols_pro"].append(gols_pro)
        df_selecao["gols_contra"].append(gols_contra)
        df_selecao["grupo"].append(grupo.split("Group ")[-1])
        df_selecao["saldo_gols"].append(gols_pro - gols_contra)
        aproveitamento = pontos / (len(df_grupo) * 3)
        df_selecao["aproveitamento"].append(round(aproveitamento, 3))

df_selecao = pd.DataFrame(df_selecao)

map_grupo = {'Round of 16': 'Oitavas de Final', 
            'Quarter-final': 'Quartas de Final', 
            'Semi-final': 'Semifinal', 
            'Third place': 'Disputa de 3Âº Lugar', 
            'Final': 'Final'}

df_selecao.replace({'grupo': map_grupo}, inplace=True)

df_selecao.head(10)
```

```python
df_selecao.to_csv('../dados/selecoes.csv', index=False)
```

# Tabela Grupos

```python
lista_grupos = list(df['category'].unique())
df_grupos = {"grupo": [],
             "partidas": []}

for grupo in lista_grupos:
    df_aux = df[df['category'] == grupo]
    patidas = []
    for index, row in df_aux.iterrows():
        patidas.append(row['team1'] + ' x ' + row['team2'])
        df_grupos["grupo"].append(grupo.split("Group ")[-1])

    df_grupos["partidas"].extend(patidas)
map_grupo = {'Round of 16': 'Oitavas de Final', 
            'Quarter-final': 'Quartas de Final', 
            'Semi-final': 'Semifinal', 
            'Third place': 'Disputa de 3Âº Lugar', 
            'Final': 'Final'}

df_grupos = pd.DataFrame(df_grupos)
df_grupos.replace({'grupo': map_grupo}, inplace=True)
df_grupos.head(10)
```

```python
df_grupos.to_csv('../dados/grupos.csv', index=False)
```

# Tabela Partida

```python
# data, hora, casa, visitante, placar_casa, placar_visitante, posse_bola_casa, 
#                     posse_bola_visitante, amarelos_casa, amarelos_visitante, vermelhos_casa, vermelhos_visitante,
#                     no_passes_casa, no_passes_visitante, no_escanteios_casa, no_escanteios_visitante

df_partidas = {"casa": [], "visitante": [], "placar_casa": [], "placar_visitante": [], "data": [], "hora": [], "posse_bola_casa": [],
                "posse_bola_visitante": [],  "grupo": [], "amarelos_casa": [], "amarelos_visitante": [], "vermelhos_casa": [], "vermelhos_visitante": [],
                "no_passes_casa": [], "no_passes_visitante": [], "no_escanteios_casa": [], "no_escanteios_visitante": []}

for index, row in df.iterrows():
    df_partidas["casa"].append(row['team1'])
    df_partidas["visitante"].append(row['team2'])
    df_partidas["placar_casa"].append(row['number of goals team1'])
    df_partidas["placar_visitante"].append(row['number of goals team2'])
    df_partidas["data"].append(row['date'])
    df_partidas["hora"].append(row['hour'])

    posse_bola_casa = float(row['possession team1'].split("%")[0]) / 100
    posse_bola_visitante = float(row['possession team2'].split("%")[0]) / 100

    df_partidas["posse_bola_casa"].append(posse_bola_casa)
    df_partidas["posse_bola_visitante"].append(posse_bola_visitante)
    
    df_partidas["grupo"].append(row['category'].split("Group ")[-1])
    df_partidas["amarelos_casa"].append(row['yellow cards team1'])
    df_partidas["amarelos_visitante"].append(row['yellow cards team2'])
    df_partidas["vermelhos_casa"].append(row['red cards team1'])
    df_partidas["vermelhos_visitante"].append(row['red cards team2'])
    df_partidas["no_passes_casa"].append(row['passes team1'])
    df_partidas["no_passes_visitante"].append(row['passes team2'])
    df_partidas["no_escanteios_casa"].append(row['crosses team1'])
    df_partidas["no_escanteios_visitante"].append(row['crosses team2'])

df_partidas = pd.DataFrame(df_partidas)
df_partidas.replace({'grupo': map_grupo}, inplace=True)
df_partidas.head(10)
```

```python
df_partidas.to_csv('../dados/partidas.csv', index=False)
```
